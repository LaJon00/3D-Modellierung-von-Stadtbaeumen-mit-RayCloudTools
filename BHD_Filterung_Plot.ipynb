```python
import numpy as np
import matplotlib.pyplot as plt
import json

# Daten aus JSON-Datei laden
with open("data_all.json", "r") as file:
    data = json.load(file)

# Extrahieren der Baumnummern und Messwerte
baum_nummern = list(data.keys())
baumkataster = [values[0] for values in data.values()]
pi_band = [values[1] for values in data.values()]
rct = [values[2] for values in data.values()]

# Indizes f√ºr die Baumnummern erstellen
indices = np.arange(len(baum_nummern))

# Scatterplot erstellen
plt.figure(figsize=(12, 6))

# Werte plotten
plt.scatter(indices, baumkataster, label="Baumkataster", color="blue", alpha=0.7)
plt.scatter(indices, pi_band, label="Pi-Band", color="green", alpha=0.7)
plt.scatter(indices, rct, label="RayCloudTools", color="red", alpha=0.7)

# Diagramm formatieren
plt.suptitle(r"$\bf{Brusth\"ohendurchmesser}$", fontsize=20, ha='center')
plt.title("Baumkataster, Pi-Band Messung und RayCloudTools", fontsize=14)
plt.xlabel("Baumnummer", fontsize=12)
plt.ylabel("BHD [cm]", fontsize=12)
plt.xticks(indices, baum_nummern, rotation=45)
plt.legend(edgecolor='black')
plt.grid(alpha=0.3)
plt.tight_layout()

# Plot speichern und anzeigen
plt.savefig("BrusthoehenDurchmesser_alle.png", dpi=300)
plt.show()
```
